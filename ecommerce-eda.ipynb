```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading data
ecommerce_df = pd.read_csv('Ecommerce Purchases')
```

```python
# top 10 rows
ecommerce_df.head(10)
```

```python
#Check Datatype of Each Column
ecommerce_df.info()
```

```python
#rename columns
new_names = {
            'AM or PM':'am_pm',
            'Address':'address',
            'Lot':'lot','Job':'job',
            'IP Address':'ip_address',
            'Language':'language',
            'Purchase Price':'purchase_price',
            'Browser Info':'browser_info',
            'Credit Card':'credit_card',
            'CC Exp Date':'cc_exp_date',
            'CC Provider':'cc_provider',
            'CC Security Code':'cc_Security_Code',
            'Email':'email',
        }

ecommerce_df = ecommerce_df.rename(columns=new_names)
ecommerce_df
```

```python
#check nan/null values
ecommerce_df.isnull().sum()
```

```python
#min,max,mean
ecommerce_df.describe(include='all')
```

```python
#am_pm frequency
frequency_am_pm = ecommerce_df['am_pm'].value_counts()
frequency_am_pm
```

```python
#am_pm frequency plot
plt.pie(frequency_am_pm,labels=['PM','AM'])
plt.title('am_pm frequency purchase')
plt.show()
```

```python
#most frequency email provider
emails_frequency = ecommerce_df['email'].apply(lambda row: row.split('@')[1]).value_counts()[:5]
emails_frequency
```

```python
#the most frequency email provider plot
plt.title('Top 5 most common email provider')
plt.pie(emails_frequency,labels=['Hotmail','Yahoo','Google','Smith','Williams'])
plt.show()
```

```python
#browse unique values
ecommerce_df.browser_info.unique()
```

```python
#The most common operating system
mac = ecommerce_df.browser_info.str.contains('Macintosh')
windows = ecommerce_df.browser_info.str.contains('Windows')
linux = ecommerce_df.browser_info.str.contains('Linux')
so_common_mac = len(ecommerce_df[mac])
so_common_windows = len(ecommerce_df[windows])
so_common_linux = len(ecommerce_df[linux])
```

```python
plt.title('The most common operating system')
plt.pie([so_common_windows,so_common_mac,so_common_linux],labels=['Windows','Mac','Linux'])
plt.show()
```

```python
#Language vs Purchase Price
plt.title("Language vs Purchase Price")
languages = ecommerce_df['language'].unique()
language_group = ecommerce_df.groupby('language')['purchase_price'].apply(list)
plt.boxplot(language_group)
plt.xticks(range(1,1+len(languages)),languages,rotation=45)
plt.xlabel("Languages") 
plt.ylabel("Purchase Price")
plt.show()
```

```python
#The most common cc provider
common_provider = ecommerce_df.cc_provider.value_counts()[:5]
common_provider
```

```python
labels_providers = ['JCB 16 digit','VISA 16 digit','JCB 15 digit','American Express','Maestro']
plt.title("Top 5 most common cc provider")
plt.pie(common_provider,labels=labels_providers)
plt.show()
```
